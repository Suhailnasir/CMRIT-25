```python
import cv2 as cv
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc('figure',figsize=(20,5)) #set figure size
```

```python
#Edges
img = cv.imread('Images/messi.jpeg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
edges = cv.Canny(img, threshold1=100, threshold2=200)
plt.imshow(edges)
```

```python
#Feature-corner
import numpy as np
img = cv.imread('Images/chessboard.png')
img_gray= cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_grayf = np.float32(img_gray) # convert to float
dest = cv.cornerHarris(img_grayf, 2,5,0.02)
dest = cv.dilate(dest,None)  #dilate for marking corners
img[dest>0.01*dest.max()] = [0,0,255]
plt.imshow(img)
```
