```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_lake = cv.imread('puppy.png')
plt.imshow(img_lake)
```

```python
img_lake = img_lake[:,:,::-1]
plt.imshow(img_lake)
```

```python
w,h,c = img_lake.shape
print(w,h,c)
print(img_lake[0,0])
```

```python
#center
cx,cy = w//2,h//2
print(cx,cy)
```

```python
tl = img_lake[0:cx,0:cy]
tr = img_lake[:cx,cy:]
bl = img_lake[cx:,:cy]
br = img_lake[cx:,cy:]
```

```python
plt.subplot(221);plt.imshow(tl)
plt.subplot(222);plt.imshow(tr)
plt.subplot(223);plt.imshow(bl)
plt.subplot(224);plt.imshow(br)
```
