```python
import cv2 as cv
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc('figure',figsize=(20,5)) #set figure size
```

```python
img = cv.imread("Images/messi.jpeg")
img_rgb = img[:,:,::-1]
avgBlur = cv.blur(img_rgb,(5,5))
gaussianBlur = cv.GaussianBlur(img_rgb,(5,5), 0)
images = [img_rgb, avgBlur, gaussianBlur]
titles = ["Original", "Average Blur", "Gaussian Blur"]
for i in range(3):
    plt.subplot(1,3,i+1); plt.title(titles[i]); plt.imshow(images[i])
```
