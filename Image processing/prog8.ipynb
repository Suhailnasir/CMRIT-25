```python
import cv2 as cv
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc('figure',figsize=(20,5)) #set figure size
```

```python
img = cv.imread('Images/puppy.png')
img_rgb = img[:,:,::-1]

#rotate
img_r90 = cv.rotate(img_rgb, cv.ROTATE_90_CLOCKWISE)
img_rc90 = cv.rotate(img_rgb, cv.ROTATE_90_COUNTERCLOCKWISE)
img_r180 = cv.rotate(img_rgb, cv.ROTATE_180)

#For displaying or plotting
plt.subplot(221); plt.title('Original'); plt.imshow(img_rgb);
plt.subplot(222); plt.title('Rotate 90'); plt.imshow(img_r90)
plt.subplot(223); plt.title('Rotate 90 cc'); plt.imshow(img_rc90)
plt.subplot(224); plt.title('Rotate 180'); plt.imshow(img_r180)
```

```python
#scaling
h,w,channels = img_rgb.shape
twox = cv.resize(img_rgb,None,fx=2,fy=2, interpolation=cv.INTER_LINEAR) #interpolation optional
halfx = cv.resize(img_rgb,(w//2,h//2), interpolation=cv.INTER_AREA)
nonUnf = cv.resize(img_rgb,None,fx=1.5,fy=0.5)
images = [img_rgb,twox,halfx,nonUnf]
titles= ['Original','2x','0.5x','non Uniform Scaling']

for i in range(4):
    plt.subplot(2,2,i+1); plt.title(titles[i]); plt.imshow(images[i]);
```

```python
#Translation
import numpy as np
tx=10; ty=50;
h,w,channels = img_rgb.shape
M = np.float32([[1,0,tx],[0,1,ty]])
trans_img = cv.warpAffine(img_rgb,M,(w,h))
plt.subplot(121); plt.title("Original"); plt.imshow(img_rgb);
plt.subplot(122); plt.title("Translated"); plt.imshow(trans_img)
```
